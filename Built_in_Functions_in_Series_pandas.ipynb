#<font color=green>**Built in functions in Series**:  
**Following shows some built in functions of the series**

**Function:** head(n)  
**Descritpion:** returns the first n members of the series.If n not specified then by default first five members are displayed.

```python
import pandas as pd
friends=pd.Series(["Rohan","Susan","James","Riya","Sumit","Abhinav","Vihaan"],
index=[11,22,33,44,55,66,77])
print(friends.head(2))
```

**Function:** tail(n)  
**Descritpion:** returns the last n members of the series.If n not specified then by default last five members are displayed.

```python
import pandas as pd
friends=pd.Series(["Rohan","Susan","James","Riya","Sumit","Abhinav","Vihaan"],
index=[11,22,33,44,55,66,77])
print(friends.tail())
```

**Function:** count(n)  
**Descritpion:** It counts the total non NaN values in a series

```python
import pandas as pd
friends=pd.Series(["Rohan","Susan","James","Riya","Sumit","Abhinav","Vihaan"],
index=[11,22,33,44,55,66,77])
print(friends.count())
```
