##<font color=orange>**Series:**</font>  
Is a one dimensional array that can store data of any type like integer,string,float,python objects,etc.  
It represents the core building block of the Pandas Data Analysis Library and can be thought of conceptually as a single column in an Excel spreadsheet or database table.


**Creating Series:**  
Different types of series are illustreated below with sample code:

```python
#Creating an empty series
import pandas as pd
Emp=pd.Series()
```

```python
#Creating a Series from a numpy array
import pandas as pd
import numpy as np
data=np.array([10,30,50])
s1=pd.Series(data)
print(s1)
```

```python
#Creating series using labelled index
import pandas as pd
friends=pd.Series(["Rohan","Susan","James"],index=[11,22,33])
print(friends)
```

```python
#Creating a Series from a Python list
import pandas as pd
cities=['Delhi','Mumbai','chennai','Kolkata']
s2=pd.Series(cities)
print(s2)
```

```python
#Creating series using labelled index
import pandas as pd
month=["June","August","October","December"]
s4=pd.Series(month,index=[6,8,10,12])
print(s4)
```

```python
#Creating series from a python dictionary
import pandas as pd
dict={6:"June",8:"August",10:"October",12:"December"}
s5=pd.Series(dict)
print(s5)
```
