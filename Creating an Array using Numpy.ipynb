#**Creating an Array using Numpy:**
Different ways to create an array using Numpy are as follows:

<font color=green> Creating one dimensional array:

```python
import numpy
rollno=numpy.array([1,2,3])
print(rollno)
```

<font color=green> Create a sequential 1D array with values as multiples of 10 from 10 to 100:

```python
import numpy as np
a=np.arange(10,101,10)
print(a)
```

<font color=green>Creating one-dimensional array with 4 randam values:

```python
import numpy as np
a=np.random.random(4)
print(a)
```

<font color=green>Creating two-dimensional array of  3 rows and 4 columns with random integer values less than 10:

```python
import numpy as np
a=np.random.randint(10,size=(3,4))
print(a)
```

<font color=green>Creating two-dimensional array of 3 rows and 4 columns with all ones as value:

```python
import numpy as np
a=np.ones((3,4))
print(a)
```

<font color=green>Creating two-dimensional array of 3 rows and 3 columns with all zeros as value:

```python
import numpy as np
a=np.zeros((3,3))
print(a)
```

<font color=green>Creating two-dimensional array of 3 rows and 4 columns with all six as value:

```python
import numpy as np
a=np.full
```

<font color=green>we can create two dimensional (2-D) arrays by passing nested lists to the array() function

```python
import numpy as np
a=np.array([[15,'Pranay',96],[16,'pranya',92]])
print(a)
```
