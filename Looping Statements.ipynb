# **Looping Statements**  
* The process of repetition of a set of instructions based on a criterion is called a loop.
* There are two types of looping statements in Python


**<font color="red">The while loop</font>**
* The for loop is used to repeat a set of instructions for a fixed number of times
* It means when the number of iterations is known/definite before we start with the execution of a loop
* It is therefore also known as a **definite loop**
* Identation of statements  is must to specify the block of statements to be repeated using **for** loop
* There are two different ways of using for loop

**Using Sequence**

```python
for count in [1,2,3,4,5]:
  print(count)
```

```python
for friends in ["Anu","Ritu","All","Sonia"]:
  print(friends)
```

```python
for vowels in "AEIOU":
  print(vowels)
```

Using **range** function

```python
for count in range(1,5,1):
  print("Welcome")
print("Iteration Finished")
```

```python
for A in range(1,11,2):
  print(A,end=",")
```

```python
for var in range(100,10,-10):
  print(var,end=",")
```

```python
for var in range(5,10,2):
  print(var,end=",")
```
