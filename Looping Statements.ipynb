# **Looping Statements**  
* The process of repetition of a set of instructions based on a criterion is called a loop.
* There are two types of looping statements in Python


##**<font color="blue">The for loop</font>**
* The for loop is used to repeat a set of instructions for a fixed number of times
* It means when the number of iterations is known/definite before we start with the execution of a loop
* It is therefore also known as a **definite loop**
* Identation of statements  is must to specify the block of statements to be repeated using **for** loop
* There are two different ways of using for loop

<font color=brown>**Using Sequence:**\
In this type of for loop,a sequence of values is used over which the loop iterate.The syntax to use the for loop with a sequence is:\
<font color=green>*for < counter variable > in < sequence >:*\
&emsp; *statements*</font>\
where,
* <font color=blue>for <font color=black>is a reserved keyword
* <font color=blue>counter variable <font color=black> can be any identifier that keeps a count of the loop
* <font color=blue>sequence <font color=black>can be any value like integer,string,list,etc
* <font color=blue>Statements <font color=black> always indented can be single or a block

**<font color=olive>For Examples:**

```python
for count in [1,2,3,4,5]:
  print(count)
```

```python
for Friends in ["Anu","Ritu","All","Sonia"]:
  print(Friends)
```

```python
for Vowels in "AEIOU":
  print(Vowels)
```

<font color=brown>**Using range function**\
The range() function is an built in function that is used to generate a sequence of values between the specified range\
The syntax to use the for loop with a range() function is:\
<font color=green>for < var > in range (< start >,< End >,< Step >):\
&emsp; < Statements > </font>\
Where
* <font color=blue>for <font color=black>is a reserved keyword
* <font color=blue>Start,end,step <font color=black>are parameters of a <font color=blue>range()</font> function and will always be **integers**
* <font color=blue>Start  <font color=black>is a starting value of loop,End is an ending value(not exclusive) of loop,and Step is the number by which counter variable will be incremented or decremented .By default step value is 1
* If only **one** parameter is used the **start** becomes 0 and **step** beomes **1** as default
* If Start > End the step=is a negative integer
* If Start < End then Step=is a positive integer
* If Start >= End and Step  value is not specified,the loop will not execute as this is an invalid condition



**For example:**

```python
for count in range(1,5,1):
  print("Welcome")
print("Iteration Finished")
```

```python
for A in range(1,11,2):
  print(A,end=",")
```

```python
for var in range(100,10,-10):
  print(var,end=",")
```

```python
for var in range(5,10,2):
  print(var,end=",")
```

## **<font color="blue">The while loop**

* The while loop is used to repeat a set of instructions as long as the condition is True.
* It means when the number of iterations is not fixed/indefinite before we start with the execution of a loop.
* It is therefore known as indefinite loop
* Identation of statements is must to specify the block of statements to be repeated using while loop
* This loop is also called as <font color=brown> an entry-controlled loop</font> as it checks for the condition in the beginning only.
* If the condition is True then the body of loop will be exceuted.
* If the conditionis False then it will not be allowed to enter within the loop and it stops

**Syntax for while loop**

<font color=green>while < condition > :\
&emsp; Statements\
&emsp; increment or decrement counter variable</font>\
where,
* <font color=blue>while</font> is a reserved keyword
* <font color=blue>condition</font> is the criterion to repeat the instructions.The instructions repeat till the condition is True.As soon as the condition is False it exits from the loop
* <font color=blue>Statements</font>are always indented can be single or a block.They are repeated till the condition is True.
Increment/decrement counter variable to change its value\
Some of the while loop are given below:

```python
count=1
while count <=5:
  print("Hello")
  count+=1
print("program ends")
```

```python
i=1
while i<6:
  print(i)
  i+=1
```

```python
#input a number abd check for even or odd
#The whole process should continue till the user wants
ans='y'
while ans=='y':
  Num=int(input("Enter a number:"))
  if Num%2==0:
    print(Num,"is an even number")
  else:
    print(Num,"is an odd number")
  ans=input("Do you want to continue(y/n):")
print("Program ends")
```

```python
#program to reverse the digits of a number
Num=int(input("Enter a number"))
Dig=0
while Num>0:
    R=Num%10
    Dig=(Dig*10)+R
    Num//=10 #Num=Num//10
print("The reverse of the number is",Dig)
```

```python
# program to add natural numbers up to sum=1+2+3+4.....+n
n=int(input("Enter the value of n: "))
sum=0
i=1
while i<=n:
  sum+=i
  i+=1
print("The sum is",sum)
```
