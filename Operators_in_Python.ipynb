# **Operators in python**  
Operators are special symbols used to perform mathematical operations,logical and relational operations on variables and values.

# **Arithmetic Operators**  
Are used to perform mathematical operations on variables and values

**Addition**
* Adds two numbers
* With string it concatenates two values

```python
10+4
```

```python
2.5+4
```

```python
"Good"+"Morning"
```

**Subtraction**  
* Subtracts second value from first value

```python
14-9
```

```python
5.5-2
```

 **Multiplication**  
 * Multiplies two values
 * Note:String * int repeats the string that many times

```python
5*4
```

```python
3.5*2
```

```python
"Bye"*2
```

 **Division**  
 * Divides two numbers and returns decimal value

```python
9/3
```

```python
6/3.0
```

```python
17/2
```

**Remainder or Modulus**  
* Returns the remainder of the division

```python
17%3
```

```python
3%5
```

**Exponential**  
* First number raised to the power second number

```python
2**5
```

```python
3.5**2
```

**Floor Division**

```python
17//3
```

**Equal to**  
* Returns True if both the values are same

```python
a=16
b=12
a==b
```

**Not equal to**  
* Returns True if both the values are same different

```python
a="Hello"
b="hello"
a!=b
```

**Greater than**

```python
a=5
b=10
a>b
```

**Less than**

```python
a=5
b=10
a<b
```
