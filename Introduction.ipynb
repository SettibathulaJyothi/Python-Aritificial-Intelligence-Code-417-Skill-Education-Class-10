**Keywords:**  
Python keywords are reserved, predefined words that have a built-in meaning.
they are restricted, you cannot use them as names for variables, functions, classes, or any other identifiers in your code

```python
import keyword
print(keyword.kwlist)
```

**Identifier**  
An identifier is a user-defined name given to a program element like a variable, dictionary,list,tuple,function, class, module, or other object in Python  
**Rules:**  
* Allowed Characters: Use only alphanumeric characters (letters A-Z, a-z, digits 0-9) and underscores (_).  
* Starting Character: Must begin with a letter or an underscore. It cannot begin with a digit.  
* Case-Sensitivity: Python is strictly case-sensitive. myVar, MyVar, and MYVAR are three completely unique identifiers.  
- No Keywords: You cannot use Python's reserved keywords (like if, else, for, class, or def) as identifiers.  
+ No Special Characters: Symbols like @, $, #, !, %, or spaces are completely forbidden

**Literals**
