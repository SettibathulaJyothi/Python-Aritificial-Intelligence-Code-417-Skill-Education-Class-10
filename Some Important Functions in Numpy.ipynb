# **Some Important Functions in Numpy**

**Type:** Type of an object i.e.array  

**Function:** type(ARR)

```python
import numpy as np
ARR=np.array((1,2,34))
print(type(ARR))
```

**Type:** Check the dimensions of an array

**Function:** ARR.ndim

```python
import numpy as np
ARR=np.array(([1,2,3,4],[3,4,5,6]))
print(ARR.ndim)
```

**Type:** Shape of an array i.e.length of the array dimensions

**Function:** ARR.shape

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.shape)
```

**Type:** Size of an array i.e.counts the number of elements.

**Function:** ARR.size

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.size)
```

**Type:** DataType of elements stored in the array

**Function:** ARR.dtype

```python
import numpy as np
ARR=np.array([[1,2,3,4],[3,4,5,6]])
print(ARR.dtype)
```

**Type:** Maximum value in the element of the array\

**Function:** ARR.max()

```python
import numpy as np
ARR=np.array([[10,34,45,23,12]])
print(ARR.max())
```

**Type:** Row wise and column wise maximum value

**Function:** ARR.max(axis=1) for row\
&emsp;&emsp;&emsp;&emsp;&emsp;ARR.max(axis=0) for column

```python
import numpy as np
ARR=np.array([[11,2,13,4],[3,4,5,6]])
print("Rowwise max: ",ARR.max(axis=1))
print("Columnwise max: ",ARR.max(axis=0))
```

**Type:** Row wise and column wise minimum value

**Function:** ARR.min(axis=1) for row\
&emsp;&emsp;&emsp;&emsp;&emsp;ARR.min(axis=0) for column

```python
import numpy as np
ARR=np.array([[11,2,13,4],[3,4,5,6]])
print("Rowwise min: ",ARR.min(axis=1))
print("Columnwise min: ",ARR.min(axis=0))
```

**Type:** Sum of all values in the given array

**Function:** ARR.sum()

```python
import numpy as np
ARR=np.array([[11,2,13,4],[3,4,5,6]])
print("Row wise sum : ",ARR.sum(axis=1))
print("Column wise sum : ",ARR.sum(axis=0))
print("sum is :",ARR.sum())
```

**Type:** Sorting the array

**Function:** ARR.sort()

```python
a=np.array([12.4,-10,23,29,15,-1,45,33,-14])
print(np.sort(a))
b=np.array([[-9,5,18,9,12],[10,11,3,-5,-10]])
print(np.sort(b,axis=1))
print(np.sort(b,axis=0))
```
